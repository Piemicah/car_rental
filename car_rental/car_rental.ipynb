# CAR RENTAL TRACKING SERVICES

### Load necessary libraries

```python
import pandas as pd
import numpy as np
```

```python
car = pd.read_csv('car_rental.csv')
```

```python
car
```

### Add column "duration-hours" for duration in hours

```python
car['duration-hours'] = (car['end timestamp']-car['start timestamp'])/3600
```

```python
car.head()
```

###  Add column "late return" to indicate true or false for late return (beyond 24 hours)

```python
car['late return']=car['duration-hours'].apply(lambda x:x>24)
```

```python
car.head()
```

### Add column "car damage" to indicate true or false if the car was damaged on return

```python
car['car damaged'] = car['comment on return'].apply(lambda x: not pd.isna(x))
```

```python
car.head()
```

### create a dataframe "summary" that shows summary for each session

```python
summary = car.drop(columns=['comment on collection','comment on return'])
```

```python
summary
```

###  convert the summary to JSON format

```python
summary_json = summary.to_json()
```
